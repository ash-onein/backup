```python
import requests
from bs4 import BeautifulSoup
import re

# Fetch YouTube page content
url = "https://www.youtube.com/watch?v=UIGth6upGdI"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Fetch title from meta tag
title = soup.find("meta", property="og:title")["content"]

# Fetch description using regex
pattern = re.compile(r'(?<=shortDescription":").*?(?=","isCrawlable)')
description_match = pattern.findall(str(soup))
description = description_match[0].replace("\\n", "\n") if description_match else "Description not found."
cleaned_description = description.split("#", 1)[0].strip()


print(f"Title: {title}")
print(f"Description: {cleaned_description}")
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Step 1: Fetch and parse the JSON data
url = "https://www.goodreturns.in/videos/tagFeeds/?tags=&limit=5000&oi_datatype=json"
response = requests.get(url)
data = response.json()  # Assuming the API returns JSON directly

# Step 2: Extract required fields
# Ensure all fields are present in the JSON data
rows = []
for item in data:
    rows.append({
        "description": item.get("description", ""), 
        "title": item.get("title", ""), 
        "title_eng": item.get("title_eng", ""),
        "published": item.get("published", ""),
        "url": item.get("url", ""),
        "wise_tags": item.get("wise_tags", "")  # Combine tags into a single string
    })


# Step 3: Create a DataFrame
df = pd.DataFrame(rows)

# Step 4: Save the table to a CSV file (optional)
df.to_csv("video_GR.csv", index=False)

# Step 5: Display the DataFrame
print(df)
```
